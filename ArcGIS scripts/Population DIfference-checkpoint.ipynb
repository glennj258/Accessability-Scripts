# Find Population Difference between 2016 and 2021 meshblocks

This finds the population difference between 2016 and 2021 ABS Census population data at the meshblock level. Due to the increased amount nd different spatial distributions of 2021 meshblocks, this cannot be done by simply subtracting/dividing the two layers. Instead, it finds the percentage increase and absolute difference in population for all 2016 mesh blocks, and gives an estimated of these for 2021 meshblock from projections of 2021 data back onto what these might have been in 2016.

This analysis relies on the fact that meshblocks are only created by splitting existing mesh blocks in two, and that no mesh blocks are lost from 2016 to 2021.

Inputs: An MB21 layer containing populations, and its corresponding 2016 meshblock and its population. (Fields: Person_21, MB_CODE16, Person_16)

Outputs: An MB21 layer containing the population percentage change and estimated absolute change from 2016.

## Function Definitions

```python
def find_population_change(pop21_df):
    '''Takes an inputs population MB21 dataframe, with the fields "MB_CODE16", "Person_21", and "Person_16".
    Outputs the dataframe with two columns added, the percentage population increase, and the absolute population increase.
    '''
    #initialise a table to put the data into
    diff_data = []
    diff_columns = ["MB_CODE21", "Pop. Percent Change", "Pop. Change", "Pop. Density Change"]
    
    #as their are less 2016 mesh blocks, we can only compare on their indices, and then project down from there
    for MB16 in pop21_df["MB_CODE16"].values:

        #find the set of all the 21 meshblocks for this 2016 meshblock 
        MB21_series = pop21_df[pop21_df["MB_CODE16"] == MB16] # not actually a series (its a df), but realised this too late


        #get the overall 2016 and 2021 populations
        pop16 = MB21_series["Person_16"].values[0]

        pop21_sum = sum(MB21_series["Person_21"].values)


        #find the percentages and percentage increases
        #percent_of_16 = 100 * pop21_sum / pop16 if pop16 != 0 else 100 #works but less clear

        if pop16 != 0: #avoid divide by 0
            percent_of_16 = 100 * pop21_sum / pop16  
            percent_increase = percent_of_16 - 100
        else:
            percent_of_16 = 0
            percent_increase = 10000 #set an arbritarily large population increase
       


        #find the absolute (non-percentage) differences
        pop_diff_MB16 = pop21_sum - pop16 #not used, but could be helpful
        
        for index, row in MB21_series.iterrows():
    
            pop21 = row["Person_21"]

            #estimate the 2016 population using the percentage change that we have
        ## TO DO: Something wrong here
            est_pop16 = int(pop21/(percent_of_16/100)) if percent_of_16 != 0 else 0

            abs_pop_diff_est = pop21 - est_pop16


            #find the population density change
            dens_diff = abs_pop_diff_est/row["AREASQKM21"]

            #add the data to the list in the correct order (MB_CODE21, percent_diff, abs_diff, abs_dens_diff)
            MB21_entry = [row["MB_CODE21"], round(percent_increase,1), abs_pop_diff_est, int(dens_diff)]

            diff_data.append(MB21_entry)
    
    #convert data to a dataframe, and merge with the input dataframe
    diff_df = pd.DataFrame(diff_data, columns = diff_columns)
    diff_df = diff_df.drop_duplicates() #the code creates a lot of duplicates, not sure why, but they are removed here
    
    #merge with original data based on mesh block ID
    pop21_change_df = pop21_df.merge(diff_df, on = "MB_CODE21")
    
    
    return pop21_change_df
        
```

### Medium density development scenario calculations

Want to raise all mesh blocks in the inner north and south (which are not already above the density to raise to) to a density required for the population increase from 2016 to 2021.

```python
pop21_r_df.columns
```

```python
#initilise a column with the 2016 population/ density  (pop densities need to be integers to enable comaprisons) 
pop21_r_df['DS_mdINIS_dev_pop_dens'] = [int(i) for i in list(pop21_r_df.Population_density_16)]
pop21_r_df['DS_mdINIS_dev_pop'] = [float(i) for i in list(pop21_r_df.pop16_est)]
```

```python
#create a new dataframe for only the inner north and south (and exclude education)
pop21_INIS_edu_df = pop21_r_df[(pop21_r_df['SA3_NAME21'] == 'North Canberra') | (pop21_r_df['SA3_NAME21'] == ('South Canberra'))]
pop21_INIS_df = pop21_INIS_edu_df[pop21_INIS_edu_df["MB_CAT21"] != "Education"]
pop21_INIS_df
```

```python
#find the list of population densities to progressively increase through
INIS_pop_densities = pop21_INIS_df['DS_mdINIS_dev_pop_dens'].sort_values().unique()
INIS_pop_densities
```

```python
pop_increase = 0
i = 0

while pop_increase < 59922:
    
    #get the lowest population density, and the population density to increase this one to
    try:
        lowest_popdens = INIS_pop_densities[i]
        next_lowest_popdens = INIS_pop_densities[i+1]
        
    except:
        print("Densities could not be found to meet population increase")
        
    print("lowest: ", lowest_popdens)
    print("next lowest: ",next_lowest_popdens)
    i += 1
    
    popdens_diff = next_lowest_popdens - lowest_popdens
    print("difference: ",popdens_diff)
    
    
    print(sorted(pop21_INIS_df['DS_mdINIS_dev_pop_dens'].values))
    
    pop_to_increase_df =  pop21_INIS_df[pop21_INIS_df['DS_mdINIS_dev_pop_dens'] == lowest_popdens]
    print(pop_to_increase_df)
                                        
    #we want to add up the population as floats, as we will be doing this incrementally, and will lose data with integers                                    
    pop_to_increase_df.DS_mdINIS_dev_pop = pop_to_increase_df.DS_mdINIS_dev_pop.astype(float)

    
    for index, row in pop_to_increase_df.iterrows():
        MB_pop_increase = pop_to_increase_df["AREASQKM21"][index] * popdens_diff
        
        pop_increase += MB_pop_increase
        
        # change values in the dataframe
        #change to edit values in actual dataframe when done for real (indices will be the same)
        pop21_INIS_df.at[index, 'DS_mdINIS_dev_pop_dens'] = next_lowest_popdens
        pop21_INIS_df.at[index, "DS_mdINIS_dev_pop"] = MB_pop_increase + pop_to_increase_df.at[index, "DS_mdINIS_dev_pop"]
        
    print("pop increase so far: ", pop_increase)
        
    if i in pop21_INIS_df['DS_mdINIS_dev_pop_dens']:
        print(i)
        
    #convert populations to integers at the end
    print(i)
    
pop21_INIS_df.DS_mdINIS_dev_pop = pop21_INIS_df.DS_mdINIS_dev_pop.astype(int)#[int(MBpop) for MBpop in pop_to_increase_df["DS_mdINIS_dev_pop_dens"].values]

```
